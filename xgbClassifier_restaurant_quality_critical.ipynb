```python
import pandas as pd
from sklearn import preprocessing, metrics, model_selection
import numpy as np
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df= pd.read_csv("Downloads/dataset/train.csv")
df_test= pd.read_csv("Downloads/dataset/test.csv")
```

```python
df.drop(df.columns[[0,1,2,3,5,10,11,12]], axis=1, inplace=True)
df_test.drop(df_test.columns[[0,1,2,3,5,10,11,12]], axis=1, inplace=True)
```

```python
df["SCORE"].fillna(df["SCORE"].mean(), inplace=True)
df_test["SCORE"].fillna(df_test["SCORE"].mean(), inplace=True)
df_test.fillna(df_test.dropna().max(),inplace =True)
df.fillna(df.dropna().max(),inplace =True)
#df=df[df['CRITICAL FLAG'] != 'Not Applicable']
```

```python
le= preprocessing.LabelEncoder()
for column in df.columns:
    if df[column].dtype == type(object):
        df[column] = le.fit_transform(df[column])
for column in df_test.columns:
    if df_test[column].dtype == type(object):
        df_test[column] = le.fit_transform(df_test[column])
        
```

```python
y=df['CRITICAL FLAG']
x=df.drop(['CRITICAL FLAG'],1)
```

```python
#clf = MLPClassifier()
#clf= GaussianNB()

# param_test1 = {
#  'max_depth':range(3,10,2),
#  'min_child_weight':range(1,6,2)
# }
# gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
#  min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
#  objective= 'multi:softmax', nthread=4, scale_pos_weight=1, seed=27), 
#  param_grid = param_test1, scoring='roc_auc',n_jobs=4,iid=False, cv=5)

clf= XGBClassifier(learning_rate =0.01,
 n_estimators=100,
 max_depth=4,
 min_child_weight=6,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 reg_alpha=0.005,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)
#clf=LogisticRegression()
#clf=RandomForestClassifier()
# gsearch1.fit(x,y)
# gsearch1.best_params_, gsearch1.best_score_
```

```python
clf.fit(x,y)
preds = clf.predict(df_test)
```

```python
sub = pd.DataFrame(preds, columns=['Critical Flag']).replace(0,'Critical').replace(2,'Not Critical').replace(1,'Not Applicable')
```

```python
final_pred=pd.concat([df_test["CAMIS"],sub], axis=1)
final_pred
```

```python
final_pred.to_csv("Downloads/submission.csv", index=False)
```
